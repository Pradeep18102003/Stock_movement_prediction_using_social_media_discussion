```python
import yfinance as yf
import pandas as pd
```

```python
sp500 = yf.Ticker('^GSPC')
data = sp500.history(start='2023-09-01', end='2024-09-01')

data.reset_index(inplace=True)
data
```

```python
#Removing the time stamp in date column

data['Date'] = data['Date'].dt.date
data
```

```python
#stock movement

data['diff'] = data['Open'].shift(-1) - data['Open']
data['Movement'] = data['diff'].apply(lambda x: 1 if x>0 else 0)
data
```

```python
stock_movement_df = data[['Date', 'Movement']]
stock_movement_df
```

```python
stock_movement_df.to_csv('stock_movement.csv', index = False)
```
